```python
import matplotlib.pyplot as plt

# 不同的核
kernels = ['rbf', 'linear', 'sigmoid', 'poly']

# 训练时间和测试时间
training_time = {
    'rbf_ovo': 57.1650,
    'rbf_ovr': 57.4323,
    'linear_ovo': 71.1389,
    'linear_ovr': 71.2944,
    'sigmoid_ovo': 88.2526,
    'sigmoid_ovr': 87.9174,
    'poly_ovo': 54.9606,
    'poly_ovr': 55.9811
}

testing_time = {
    'rbf_ovo': 33.09646,
    'rbf_ovr': 33.18965,
    'linear_ovo': 13.28105,
    'linear_ovr': 13.33015,
    'sigmoid_ovo': 28.84231,
    'sigmoid_ovr': 28.89204,
    'poly_ovo': 12.78422,
    'poly_ovr': 13.06026
}

# 绘制折线图
plt.figure(figsize=(10, 6))

# 绘制训练时间折线图
plt.subplot(211)
for kernel in kernels:
    plt.plot([kernel+'+ovo', kernel+'+ovr'], [training_time[kernel+'_ovo'], training_time[kernel+'_ovr']], marker='o', linestyle='-', label=kernel)

plt.title('Training Time for Different SVM Kernels')
plt.xlabel('SVM Kernels and OVO/OVR')
plt.ylabel('Training Time (seconds)')
plt.legend()

# 绘制测试时间折线图
plt.subplot(212)
for kernel in kernels:
    plt.plot([kernel+'+ovo', kernel+'+ovr'], [testing_time[kernel+'_ovo'], testing_time[kernel+'_ovr']], marker='x', linestyle='--', label=kernel)

plt.title('Testing Time for Different SVM Kernels')
plt.xlabel('SVM Kernels and OVO/OVR')
plt.ylabel('Testing Time (seconds)')
plt.legend()

# 调整布局
plt.tight_layout()

# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 不同的核
kernels = ['rbf', 'linear', 'sigmoid', 'poly']

# 训练时间和测试时间
training_time_ovo = [57.1650, 71.1389, 88.2526, 54.9606]
training_time_ovr = [57.4323, 71.2944, 87.9174, 55.9811]

testing_time_ovo = [33.09646, 13.28105, 28.84231, 12.78422]
testing_time_ovr = [33.18965, 13.33015, 28.89204, 13.06026]

x = np.arange(len(kernels))  # 柱状图的 x 坐标

# 绘制训练时间和测试时间折线图
plt.figure(figsize=(10, 6))

# 训练时间
plt.plot(x, training_time_ovo, marker='o', linestyle='-', label='OVO Training Time')
plt.plot(x, training_time_ovr, marker='x', linestyle='--', label='OVR Training Time')

# 测试时间
plt.plot(x, testing_time_ovo, marker='s', linestyle='-', label='OVO Testing Time')
plt.plot(x, testing_time_ovr, marker='^', linestyle='--', label='OVR Testing Time')

plt.xticks(x, kernels)  # 设置 x 轴刻度
plt.title('Training and Testing Time for Different SVM Kernels')
plt.xlabel('SVM Kernels')
plt.ylabel('Time / S')
plt.legend()
plt.savefig("./demo.jpg")
# 显示图形
plt.show()
```
